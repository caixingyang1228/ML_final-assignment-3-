```python
import pandas as pd   #import many of API functions
from sklearn.cluster import KMeans
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
import numpy as np
from sklearn.metrics import precision_score, accuracy_score, confusion_matrix, recall_score, f1_score
import matplotlib.pyplot as plt
```

```python
data = pd.read_json("food_enforcement_US_bacteria.json")#  Using the API pandas to read json file
print("data = ", data.head(10))   #look through the original data 

data.to_csv("food_enforcement_US_bacteria.csv")   # save the original data imto csv file 
```

```python
data = pd.read_csv("food_enforcement_US_bacteria.csv")  #Read the original data 
data.head(5)  #observe at the top 5 rows of dataset
```

```python
feature = ["event_id", "classification", "center_classification_date", "report_date", "recall_initiation_date","recall_number",
"city",  "distribution_pattern",  "recalling_firm", "state", "reason_for_recall" , "initial_firm_notification",
   "status", "product_description",  "code_info"  ,  "address_1", "complete_address"    ]   # choose the features 
```

```python
data = data[feature].dropna(axis=0)  # clear  the nan data 
data = data.reset_index( drop = True)  # clear  the nan data 
label = data["classification"]     #choose lable data of "classification "
ID = data["event_id"]            #choose the event_id as ID data 
```

```python
for i in range(len(label)):
    if label.iloc[i] == "Class II": # digitalize the class 
        label.iloc[i] = 0
    elif label.iloc[i] == "Class I":# digitalize the class 
        label.iloc[i] = 1
    else:
        label.iloc[i] = 2     # digitalize the other class 
```

```python
def vector(data):  #vectoring the characters 
    store = []
    vectorizer = TfidfVectorizer() #use the Vectorizer API
    for i in range(len(data)):
        tfidf_model = TfidfVectorizer().fit(data.iloc[i])
        features = tfidf_model.transform(data.iloc[i])
        store2 = np.array(features.todense()).tolist()[0]
        while len(store2) < 110:
            store2.append(0)
        store.append(store2)

    return store
# vectorizer = TfidfVectorizer()


features = vector(data[["reason_for_recall"]])  #cconfirm the whole reason_for_recall were ulter into number 
dataframe = pd.DataFrame(features)
```

```python
km = KMeans(n_clusters=3 )  # invoking the Kmeans API
# km.fit(dataframe)
km.fit(features)            # fit the features 

print("label = ", km.labels_)  # print the labels  
print('KMeans precision scoreï¼',precision_score(label,km.labels_, average='micro'))  # print the precision_score  
print('KMeans Confusion Matrixï¼',confusion_matrix(label,km.labels_))   # print the confusion_matrix 
print('KMeans accuracy scoreï¼',accuracy_score(label,km.labels_))      # print the accuracy_score  
print('KMeans recall scoreï¼',recall_score(label,km.labels_,  average="micro", sample_weight=None))# print the recall_score 
print('KMeans f1 scoreï¼',f1_score(label,km.labels_, pos_label=1,average = "macro")) # print the f1_score
```

```python
plt.figure()
color = ["red", "green", "blue"]  #set the color content 
for i in range(len(ID)):
    plt.scatter(i, ID[i], color = color[label[i]]) #plot the scatter figure 
    if i == len(ID)-1:
        plt.scatter(i, ID[i], color=color[label[i]], label = "real")
plt.title("original cluster result")
plt.legend()


plt.figure()
for i in range(len(km.labels_)):
    plt.scatter(i, ID[i], color = color[km.labels_[i]])
    if i == len(km.labels_)-1:
        plt.scatter(i, ID[i], color= color[km.labels_[i]], label = "kmeans result")
plt.title("Kmeans cluster result")  #plot the scatter figure 
plt.legend()
plt.show()
```
